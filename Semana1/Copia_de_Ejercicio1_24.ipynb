```python
import pandas as pd
```

```python
# Crear un dataframe que tenga 3 registros(filas) cada una debe contener el nombre de una cancion, el artista que la canta, y la duracion en segundos
```

```python
nombre_Cancion= pd.Series(['Weekend','life','Enjoy'])
Artista = pd.Series(['Bumbury','Lenon','Elvis'])
Duracion_seg = pd.Series([1,2,3])

#print(nombre_Cancion)
#print(Artista)
#print(Duracion_seg)

df_songs = pd.DataFrame({'Nombre Cancion': nombre_Cancion, 'Artista': Artista, 'Duracion(s)': Duracion_seg})
df_songs
```

```python
# Imprimir la cola del dataframe
df_songs.tail()
```

```python
# Acceder a la columna duracion
df_songs['Duracion(s)']
```

```python
# Calcular el promedio de las duraciones
promedio = df_songs['Duracion(s)'].mean()
print(promedio)


```

```python
# Obtener los nombres de las canciones que tengan duracion mayor que el promedio

df_songs['Duracion(s)'].apply(lambda T:T>promedio)

validacion = df_songs['Duracion(s)'].apply(lambda T:T>promedio)

##df_songs['Validacion'] = pd.Series(df_songs['Duracion(s)'].apply(lambda T:T>promedio))
df_songs['Nombre Cancion'][validacion == True]

```

```python
# Agregar una columna al dataframe que contenga True si la duracion es mayor que 3 minutos
# y false si es menor
df_songs['Validacion_T_or_F'] = pd.Series(df_songs['Duracion(s)'].apply(lambda T:T>promedio))
df_songs
```

```python
# Convertir el dataframe a numpy array
import numpy  as np
array = df_songs.to_numpy()
array
```
